### **Introduction au clustering**


#### *Qu’est-ce que le clustering ?*

**Le clustering** est l'une des techniques d'analyse de données exploratoires les plus couramment utilisées pour obtenir une intuition sur la structure des données. Cela peut être défini comme la tâche d'identifier des sous-groupes dans les données de sorte que les points de données dans le même sous-groupe (cluster) soient très similaires tandis que les points de données dans différents clusters sont très différents. En d'autres termes, nous essayons de trouver des sous-groupes homogènes dans les données de sorte que les points de données de chaque groupe soient aussi similaires que possible selon une mesure de similarité telle que la distance euclidienne ou la distance basée sur la corrélation. Le choix de la mesure de similarité à utiliser dépend de l'application.

Contrairement à l'apprentissage supervisé, le clustering est considéré comme une méthode d'apprentissage non supervisé car nous n'avons pas la vérité de terrain pour comparer la sortie de l'algorithme de clustering aux véritables étiquettes afin d'évaluer ses performances. Nous voulons seulement essayer d'étudier la structure des données en regroupant les points de données en sous-groupes distincts.

#### Mesures de similarités

Une bonne méthode de clustering est une méthode qui maximise la ressemblance entre les données à l’intérieur de chaque cluster, et minimise la ressemblance entre les données des clusters différents. C’est pourquoi les résultats d’une technique de clustering dépendent fortement de la mesure de similarité choisie par son concepteur, qui doit la choisir avec prudence. En effet la mesure de similarité repose sur le calcul de la distance entre deux données, sachant que chaque donnée est composée d’un ensemble d’attributs numériques et/ou catégoriels. Plus la distance est importante, moins similaires sont les données et vice versa. Soit $x_{i}$ et $x_{j}$ deux données différentes dont on veut calculer la distance. Cette distance est composée d’une part de la distance entre les valeurs des attributs numériques et d’une autre part de la distance entre les valeurs des attributs catégoriels ou symboliques, en prenant bien sur en considération le poids (le nombre) de chaque type d’attributs.

Pour mesurer la distance entre les données à attributs numériques, plusieurs formules existent :
 - La distance euclidienne : $D_{n}(x_{i}, x_{j}) = \sqrt{\sum_{k=1}^{n_{n}}}{(x_{ik}-x_{jk})^{2}}$
 
 - La distance de Manhattan : $D_{n}(x_{i},x_{j}) = \sum_{k=1}^{n_{n}}{|x_{ik}-x_{jk}|}$

 - La distance de Minkowksi : $D_{np}(x_{i}-x_{j}) = (\sum_{k=1}^{n_{n}}{(x_{ik}-x_{jk}})^{2})^{2}$

 Il faut faire attention lors du calcul de ces distances à la normalisation des attributs, puisque les intervalles de variances des attributs peuvent être très différents, ce qui peut entraîner la dominance d’un ou de quelques attributs sur le résultat. Il est conseillé donc, de normaliser tous les attributs sur le même intervalle puis calculer la distance.

#### **Applications du Clustering**

**Le clustering** est utilisé dans la segmentation du marché, où l'on essaie de trouver des clients qui se ressemblent que ce soit en termes de comportements ou d'attributs, de segmentation/compression d'images, où nous essayons de regrouper des régions similaires, de regrouper des documents en fonction de sujets, etc.

#### 1. *Clustering basé sur des partitions*

#### 1.1 *Les $k$-means*

**L'algorithme des $k$-means** est un algorithme itératif qui tente de partitionner l'ensemble de données en $k$ sous-groupes (clusters) distincts et non superposés de manière prédéfinis, où chaque point de données appartient à un seul groupe . Il essaye de rendre les points de données intra-cluster aussi **similaires** que possible tout en gardant les clusters aussi différents (c'est à dire aussi loin) que possible. **Il attribue des points de données à un cluster de sorte que la somme de la distance au carré entre les points de données et le centroïde du cluster (qui est la moyenne arithmétique de tous les points de données appartenant à ce cluster) soit au minimum**. Moins nous avons de variation au sein des clusters, plus les points de données sont homogènes (c'est à dire similaires) au sein du même cluster.

La mise en place de l'algorithme $k$-means se fait en trois étapes et de la manière suivantes :
1. Spécifiez le nombre de clusters $k$
2. Initialisez les centroïdes en mélangeant d'abord l'ensemble de données, puis en sélectionnant au hasard $k$ points de données pour les centroïdes sans remplacement. 
3. Continuez à itérer jusqu'à ce qu'il n'y ait aucun changement dans les centroïdes. c'est-à-dire que l'affectation des points de données aux clusters ne change pas.
 - Calculez la somme de la distance au carré entre les points de données et tous les centroïdes.
 - Attribuez chaque point de données au cluster le plus proche (centre de gravité).
 - Calculez les centroïdes des clusters en prenant la moyenne de tous les points de données appartenant à chaque cluster.

Eléments important à noter dans l'exécution de l'algorithme des $k$-means

- Étant donné que les algorithmes de clustering, y compris $k$-means, utilisent des mesures basées sur la distance pour déterminer la similarité entre les points de données, il est recommandé de normaliser les données pour avoir une moyenne de zéro et un écart type de un, car presque toujours les entités d'un ensemble de données auraient différentes unités de mesures, comme l'âge ou le revenu.

- Compte tenu de la nature itérative de $k$-means et de l'initialisation aléatoire des centroïdes au début de l'algorithme, différentes initialisations peuvent conduire à différents clusters, car l'algorithme $k$-means peut rester bloqué dans un optimum local et ne pas converger vers l'optimum global . Par conséquent, il est recommandé d'exécuter l'algorithme en utilisant différentes initialisations de centroïdes et de choisir les résultats de l'exécution qui ont donné la somme inférieure de la distance au carré.

#### *Applications des $k$-means*
L'algorithme $k$-means est très populaire et utilisé dans une variété d'applications telles que la segmentation du marché, le regroupement de documents, la segmentation et la compression d'images, etc. L'objectif généralement lorsque nous effectuons une analyse de cluster est soit :
1. Obtenir une intuition significative de la structure des données que nous traitons.
2. Regroupez puis prédisez où différents modèles seront construits pour différents sous-groupes si nous pensons qu'il existe une grande variation dans les comportements des différents sous-groupes. Un exemple de cela consiste à regrouper les patients en différents sous-groupes et à construire un modèle pour chaque sous-groupe afin de prédire la probabilité du risque de crise cardiaque.

#### *Méthodes d'évaluation*
L'analyse de clustering n'a pas de métrique d'évaluation solide que nous pouvons utiliser pour évaluer le résultat de différents algorithmes de clustering. De plus, puisque $k$-means nécessite $k$ comme entrée et ne l'apprend pas à partir des données, il n'y a pas de bonne réponse en termes de nombre de clusters que nous devrions avoir dans n'importe quel problème. Parfois, la connaissance du domaine et l'intuition peuvent aider, mais ce n'est généralement pas le cas. Dans la méthodologie de prédiction de cluster, nous pouvons évaluer la performance des modèles en fonction de différents clusters $k$ puisque les clusters sont utilisés dans la modélisation en aval.

De ce fait, Deux metriques peuvent nous donner une intuition sur le choix de $k$ : 
 - La methode du coude
 - La méthode des silhoutes 

1. *La Méthode du coude*

La méthode du coude nous donne une idée de ce qu'un bon nombre $k$ de clusters serait basé sur la somme de la distance au carré (SSE) entre les points de données et les centroïdes de leurs clusters attribués. Nous sélectionnons $k$ à l'endroit où SSE commence à s'aplatir et à former un coude. Nous utiliserons l'ensemble de données de geyser et évaluerons SSE pour différentes valeurs de $k$ et verrons où la courbe pourrait former un coude et s'aplatir.

```python
# Run the Kmeans algorithm and get the index of data points clusters
sse = []
list_k = list(range(1, 6))

for k in list_k:
    km = KMeans(n_clusters=k)
    km.fit(X_std)
    sse.append(km.inertia_)

# Plot sse against k
plt.figure(figsize=(6, 4))
plt.plot(list_k, sse, '-o')
plt.xlabel(r'Number of clusters k')
plt.ylabel('Sum of squared distance');
```

Le graphique ci-dessus montre que $k=2$ est le meilleur choix. Parfois, il est encore difficile de déterminer un bon nombre de clusters à utiliser car la courbe diminue de manière monotone et peut ne pas montrer de coude ou avoir un point évident où la courbe commence à s'aplatir.

2. *Analyse des silhouettes*

L'analyse de silhouette peut être utilisée pour déterminer le degré de séparation entre les clusters. Pour chaque échantillon. Pour cela on 

- Calcule la distance moyenne de tous les points de données dans le même cluster $(a^{i})$.
- Calcule la distance moyenne à partir de tous les points de données dans le cluster le plus proche $(b^{i})$.
- Calcule le coefficient :

$$\frac{b^{i}-a^{i}}{\max(a^{i}, b^{i})}$$

Le coefficient peut prendre des valeurs dans l'intervalle $[-1, 1]$.

 - Si c'est 0, alors l'échantillon est très proche des grappes voisines.
 - si c'est alors, l'échantillon est éloigné des grappes voisines.
 - Si c'est -1 alors l'échantillon est affecté aux mauvais clusters.

Par conséquent, nous voulons que les coefficients soient aussi grands que possible et proches de 1 pour avoir de bons clusters.

```python
for i, k in enumerate([2, 3, 4]):
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(X_std)
    centroids = km.cluster_centers_
    silhouette_vals = silhouette_samples(X_std, labels)
    
print(silhouette_vals)
```

```python
for i, k in enumerate([2, 3, 4]):
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(12, 7)
    
    # Run the Kmeans algorithm
    km = KMeans(n_clusters=k)
    labels = km.fit_predict(X_std)
    centroids = km.cluster_centers_

    # Get silhouette samples
    silhouette_vals = silhouette_samples(X_std, labels)

    # Silhouette plot
    y_ticks = []
    y_lower, y_upper = 0, 0
    for i, cluster in enumerate(np.unique(labels)):
        cluster_silhouette_vals = silhouette_vals[labels == cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
        ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
        y_lower += len(cluster_silhouette_vals)

    # Get the average silhouette score and plot it
    avg_score = np.mean(silhouette_vals)
    ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])
    ax1.set_xlabel('Silhouette coefficient values')
    ax1.set_ylabel('Cluster labels')
    ax1.set_title('Silhouette plot for the various clusters', y=1.02);
    
    # Scatter plot of data colored with labels
    ax2.scatter(X_std[:, 0], X_std[:, 1], c=labels)
    ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r', s=250)
    ax2.set_xlim([-4, 4])
    ax2.set_xlim([-2, 2])
    ax2.set_xlabel('Eruption time in mins')
    ax2.set_ylabel('Waiting time to next eruption')
    ax2.set_title('Visualization of clustered data', y=1.02)
    ax2.set_aspect('equal')
    plt.tight_layout()
    plt.suptitle(f'Silhouette analysis using k = {k}',
                 fontsize=16, fontweight='semibold', y=1.05);
```

Comme le montrent les graphiques ci-dessus, n_clusters = 2a est le meilleur score de silhouette moyen d'environ 0,75 et tous les clusters étant au-dessus de la moyenne montrent qu'il s'agit en fait d'un bon choix. De plus, l'épaisseur du tracé de la silhouette donne une indication de la taille de chaque cluster. Le graphique montre que le groupe 1 a presque le double d'échantillons que le groupe 2. Cependant, lorsque nous sommes passés n_clusters à 3 et 4, le score de silhouette moyen a considérablement diminué pour atteindre environ 0,48 et 0,39 respectivement. De plus, l'épaisseur du tracé de la silhouette a commencé à montrer de grandes fluctuations. L'essentiel est le suivant : Bon n_clusters aura un score moyen de silhouette bien supérieur à 0,5 et tous les clusters auront un score supérieur à la moyenne.

## Mesure d'évaluation supervisée

```python
from sklearn.metrics import fowlkes_mallows_score
labels = [0, 0, 0, 1, 1, 1]
labels_pred = [1, 1, 2, 2, 3, 3]

FMI = fowlkes_mallows_score(labels, labels_pred)
print(f"FMI : {FMI}")
```

```python
from sklearn.metrics import (
    mutual_info_score,
    normalized_mutual_info_score,
    adjusted_mutual_info_score,
)
labels = [0, 0, 0, 1, 1, 1]
labels_pred = [1, 1, 2, 2, 3, 3]

MI = mutual_info_score(labels, labels_pred)
NMI = normalized_mutual_info_score(labels, labels_pred)
AMI = adjusted_mutual_info_score(labels, labels_pred)
print(f"MI : {MI}, NMI : {NMI} and AMI : {AMI}")
```

```python
from sklearn.metrics import (
    homogeneity_score,
    completeness_score,
    v_measure_score,
)
labels = [0, 0, 0, 1, 1, 1]
labels_pred = [1, 1, 2, 2, 3, 3]

HS = homogeneity_score(labels, labels_pred)
CS = completeness_score(labels, labels_pred)
V = v_measure_score(labels, labels_pred, beta=1.0)
print(f"HS : {HS}, CS : {CS}, V : {V}")
```
