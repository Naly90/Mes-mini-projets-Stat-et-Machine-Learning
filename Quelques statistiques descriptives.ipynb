# Quelques caractéristiques de tendance centrale et de dispersion

### Importation des packages 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from tabulate import tabulate 
%matplotlib inline 
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

### Calcul des caractéristiques de tendances centrales

```python
summary = iris.describe()
print(tabulate(summary, headers="keys", tablefmt = "fancy_grid", numalign="center"))
```

### Caractéristiques dispersion à partir des boxplots

```python
#création de boxplots univariés de chaque categories de fleures
fig, ax = plt.subplots(1, 4, figsize = (16, 6))

for i , col in enumerate(iris.columns[:-1]):
    sns.boxplot(y = col, data = iris, ax = ax[i])
    ax[i].set_title(f"Boxplot de {col}")
    
plt.tight_layout();
```

```python
#boxplot pour chaque modalité de catégorie de fleures 
fig, ax = plt.subplots(2, 2, figsize = (12, 6))

for i, col in enumerate(iris.columns[:-1]):
    sns.boxplot(x = "species", y = col, data = iris, ax = ax[i//2, i%2])
    
plt.tight_layout();
```
